```python
#Handling categorical feature Gender
df1['Gender']=df1['Gender'].map({'F':0,'M':1}) #Dictionaries are used here
#Here, the 'Gender' column has been transformed from categorical values ('F' and 'M') to numerical values (0 and 1). This is often done to prepare the data for machine learning models, which typically require numerical input.
```

```python
##fixing categorical city_category
df1_city=pd.get_dummies(df1['City_Category'],drop_first=True)
##the pd.get_dummies function is used to convert categorical variables into a format that can be provided to machine learning algorithms to do a better job in prediction. This is called "one-hot encoding". Each category value is converted into a new column and assigned a binary value of 1 or 0.
##Original Column: 'City_Category' has three categories: 'A', 'B', and 'C'.
##Without drop_first: Creates three columns (one for each category) with binary values indicating the presence of each category.
##With drop_first=True: Drops the first category ('A') and creates columns only for the remaining categories ('B' and 'C').
 ## Result Interpretation:

##If a row has 'A' in the original column, both 'B' and 'C' columns will be 0.
##If a row has 'B' in the original column, the 'B' column will be 1 and 'C' will be 0.
##If a row has 'C' in the original column, the 'C' column will be 1 and 'B' will be 0.                                                                        
```
